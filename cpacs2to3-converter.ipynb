# Using the cpacs2to3 Converter online

You can use the cpacs2to3 converter directly using this jupyter notebook interactively. Do so so, follow the following link:
https://mybinder.org/v2/gh/joergbrech/tigl-tutorials/master. Or if you are already there, navigate to your Jupyter Home website.

First, upload the file you would like to convert by pressing the ``Upload`` button at the top right. Confirm by pressing the ``Upload`` button next to the file you wish to upload.


## Conversion in this notebook ##

Conversion is as simple as

```bash
cpacs2to3 CPACSfiles/CPACS2/simple_test_guide_curves.xml -o simpletest_guidecurves_v3.xml
```

If you navigate to your Jupyter Home, you will find the converted file ``simpletest_guidecurves_v3.xml`` ready for your download:


Select it, press the ``Download`` button and presto - you have your CPACS 3 file!

## Conversion in a Terminal

BTW, it might be a bit more convenient to just start a new Terminal in your Jupyter Home and enter the command

    cpacs2to3 CPACSfiles/CPACS2/simple_test_guide_curves.xml -o simpletest_guidecurves_v3.xml
    
there instead of using this notebook.
